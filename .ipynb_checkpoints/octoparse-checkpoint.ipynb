```python
# import libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
# load data into dataframe 
rental = pd.read_csv("/Users/nikitha/Desktop/CAPSTONE/trial_final.csv")
rental.head()
```

```python
rental.info()
```

```python
# check for missing values
rental.isnull().sum()
```

```python
# drop missing values
rental = rental.dropna()
rental = rental.reset_index(drop=True)
```

```python
# check for missing values again
rental.isnull().sum()
```

```python
print("There are {} duplicate values.".format(rental.duplicated().sum()))
rental[rental.duplicated(keep=False)].head(10)
```

```python
# remove duplicate values
rental = rental.drop_duplicates()
rental = rental.reset_index(drop=True)
```

```python
# check for duplicate values after removing duplicates
print("There are {} duplicate values.".format(rental.duplicated().sum()))
rental.head()
```

```python
plt.figure(figsize=(16, 6))
correlation_mat=rental.corr()
sns.heatmap(rental.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
plt.show()
```

```python
# select columns that we want to work with
rental = rental[["Rooms", "Bathrooms", "SquareFeet","Total Floor","Gym","Furnishing","Parking","Locality","Direction","Rent"]]
rental.head(10)
```

```python
rental.info()
```

```python

# one hot encoding
rental_df = pd.get_dummies(rental)
rental_df.head()
```

```python
rental_df.info()
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(rental_df['Rent'])# Remove the labels from the features
# axis 1 refers to the columns
features= rental_df.drop('Rent', axis = 1)# Saving feature names for later use
feature_list = list(features.columns)# Convert to numpy array
features = np.array(features)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

# Random Forest

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 5, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'rupees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Fitting the Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 500, random_state = 0)
rf.fit(train_features, train_labels)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'rupees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


# Predicting R2 Score the Train set results
y_pred_rf_train = rf.predict(train_features)
r2_score_rf_train = r2_score(train_labels, y_pred_rf_train)

# Predicting R2 Score the Test set results
y_pred_rf_test = rf.predict(test_features)
r2_score_rf_test = r2_score(test_labels, y_pred_rf_test)

# Predicting RMSE the Test set results
rmse_rf = (np.sqrt(mean_squared_error(test_labels, y_pred_rf_test)))

print('R2_score (train): ', r2_score_rf_train)
print('R2_score (test): ', r2_score_rf_test)
print("RMSE: ", rmse_rf)
```
