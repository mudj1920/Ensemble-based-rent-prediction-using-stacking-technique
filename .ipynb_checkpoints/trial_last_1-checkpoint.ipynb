```python
# take a look at the statistics of the Rent column
with pd.option_context('float_format', '{:f}'.format): print(rental["Rent"].describe())
```

```python
#detecting and removing outliers
Q1 = np.percentile(rental['Rent'], 25, interpolation = 'midpoint')
 
Q3 = np.percentile(rental['Rent'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", rental.shape)
 
# Upper bound
upper = np.where(rental['Rent'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(rental['Rent'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
rental.drop(upper[0], inplace = True)
rental.drop(lower[0], inplace = True)
 
print("New Shape: ", rental.shape)
```

```python
# take a look at the statistics of the Rent column
with pd.option_context('float_format', '{:f}'.format): print(rental["Rent"].describe())
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = rental.iloc[:,[0,1,2,3,4,5,7,10,11]]
y = rental.iloc[:,-1]    #target column i.e price range#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=8)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Feature','Score']  #naming the dataframe columns
print(featureScores.nlargest(8,'Score'))
```

```python
# select columns that we want to work with
rental = rental[["Deposit", "SquareFeet", "Total floor","Total floor","Floor","Bathrooms","Rooms","Age","Longitude","Lift","Negotiability","Maintenance","Gym","Furnishing","Parking","Direction","Locality","Rent"]]
rental.head(10)
```

```python
# one hot encoding
rental_df = pd.get_dummies(rental)
rental_df.head()
```

```python
rental_df.info()
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(rental_df['Rent'])# Remove the labels from the features
# axis 1 refers to the columns
features= rental_df.drop('Rent', axis = 1)# Saving feature names for later use
feature_list = list(features.columns)# Convert to numpy array
features = np.array(features)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 25)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.tree import DecisionTreeRegressor
regressor_dt = DecisionTreeRegressor(random_state = 0)
```

```python
# Train the model on training data
regressor_dt.fit(train_features, train_labels);
```

```python
# Use the dt's predict method on the test data
predictions = regressor_dt.predict(test_features)
```

```python
# Calculate the absolute errors
errors = abs(predictions - test_labels)
```

```python
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'rupees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
```

```python
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


# Predicting R2 Score the Train set results
y_pred_dt_train = regressor_dt.predict(train_features)
r2_score_dt_train = r2_score(train_labels, y_pred_dt_train)

# Predicting R2 Score the Test set results
y_pred_dt_test = regressor_dt.predict(test_features)
r2_score_dt_test = r2_score(test_labels, y_pred_dt_test)

# Predicting RMSE the Test set results
rmse_dt = (np.sqrt(mean_squared_error(test_labels, y_pred_dt_test)))

print('R2_score (train): ', r2_score_dt_train)
print('R2_score (test): ', r2_score_dt_test)
print("RMSE: ", rmse_dt)
```
